```python
# ðŸ”¹ Instalar dependÃªncia
!pip install deepface

import requests
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace

# ðŸ”¹ Baixar uma imagem de exemplo
url = "https://raw.githubusercontent.com/serengil/deepface/master/tests/dataset/img1.jpg"
img_path = "test.jpg"
with open(img_path, "wb") as f:
    f.write(requests.get(url).content)

# ðŸ”¹ Executar anÃ¡lise
result = DeepFace.analyze(img_path=img_path, actions=["age","gender","race","emotion"])
print(result)

# ðŸ”¹ Mostrar imagem com marcaÃ§Ã£o
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# DeepFace tambÃ©m pode detectar faces e devolver bounding boxes
detections = DeepFace.extract_faces(img_path=img_path, detector_backend="opencv")

for det in detections:
    x, y, w, h = det["facial_area"].values()
    cv2.rectangle(img, (x, y), (x+w, y+h), (0,255,0), 2)

plt.imshow(img)
plt.axis("off")
plt.show()

```

```python
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace

# Rodar anÃ¡lise e detectar rostos
result = DeepFace.analyze(img_path=img_path, actions=["age","gender","emotion"], detector_backend="opencv")
detections = DeepFace.extract_faces(img_path=img_path, detector_backend="opencv")

img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

for i, det in enumerate(detections):
    region = det["facial_area"]
    x, y, w, h = region["x"], region["y"], region["w"], region["h"]

    # Caixa verde
    cv2.rectangle(img, (x, y), (x+w, y+h), (0,255,0), 2)

    # Pegar infos do resultado correspondente
    face_res = result[i] if isinstance(result, list) else result
    age = face_res["age"]
    gender = face_res["dominant_gender"]
    emotion = face_res["dominant_emotion"]

    label = f"{gender}, {age} anos, {emotion}"
    cv2.putText(img, label, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,255,0), 2)

plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
# Salvar imagem com marcaÃ§Ãµes para comprovar resultado
output_path = "results/prediction_demo.jpg"
cv2.imwrite(output_path, cv2.cvtColor(img, cv2.COLOR_RGB2BGR))
print(f"Imagem salva em {output_path}")
```
